```python
import os
import pandas as pd
from tqdm import tqdm  # 导入 tqdm 库

# 设置文件夹路径
folder_path = r'C:\Users\86155\Downloads\土库曼斯坦'

# 获取文件夹中的所有Excel文件
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx') or f.endswith('.xls')]

# 使用 tqdm 包裹文件列表，显示进度条
for filename in tqdm(excel_files, desc="Processing Excel files", unit="file"):
    file_path = os.path.join(folder_path, filename)
    
    # 读取Excel文件
    df = pd.read_excel(file_path, sheet_name=None)  # sheet_name=None 会读取所有sheet
    
    # 遍历每个sheet
    for sheet_name, sheet_data in df.items():
        # 在每个sheet中添加一列，列名为'FileName'，值为文件名
        sheet_data['FileName'] = filename
        
        # 保存修改后的表格到新的Excel文件中
        output_path = os.path.join(folder_path, f"{filename}")
        with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
            sheet_data.to_excel(writer, sheet_name=sheet_name, index=False)
    
    print(f"文件 {filename} 已处理并保存。")
```

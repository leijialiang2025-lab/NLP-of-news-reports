```python
import pandas as pd

# 假设您的Excel文件路径是excel_file
excel_file = r"D:\学习\Naraka\中国外交部.xlsx"
sheet_name = 'Sheet1'
df = pd.read_excel(excel_file, sheet_name=sheet_name)

df['字段1_文本'] = df['字段1_文本'].str.split().str[-1]
```

```python
df
```

```python
df['时间'] = df['时间'].str.replace(r'发布时间：', '', regex=True)
```

```python
df
```

```python
# 将日期列转换为datetime对象
df['时间'] = pd.to_datetime(df['时间'])

# 将datetime对象格式化为所需的字符串格式
df['时间'] = df['时间'].dt.strftime('%Y年%m月%d日')
```

```python
df
```

```python
df.to_excel(r"D:\学习\Naraka\中国外交部.xlsx", index=False)
```

```python
import pandas as pd

# 假设您的Excel文件路径是excel_file
excel_file = r"D:\学习\Naraka\巴哈马中国网.xlsx"
df = pd.read_excel(excel_file)

# 假设您要去除重复值的列名是'column_name'
# 使用drop_duplicates()方法去除重复值，只保留第一次出现的行
df = df.drop_duplicates(subset='字段1_文本', keep='first')

df['时间'] = pd.to_datetime(df['时间'])
# 将datetime对象格式化为所需的字符串格式
df['时间'] = df['时间'].dt.strftime('%Y年%m月%d日')
# 显示修改后的DataFrame
print(df)

# 如果需要，保存修改后的DataFrame回Excel文件
df.to_excel(r"D:\学习\Naraka\巴哈马中国网.xlsx", index=False)
```

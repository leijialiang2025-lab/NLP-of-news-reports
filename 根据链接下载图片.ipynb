```python
import os
import requests
from openpyxl import load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils import get_column_letter

# 定义图片保存的文件夹
output_folder = r'C:\Users\86155\Desktop\学习\加勒比地区图库'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 指定Excel文件的路径
excel_path = r'C:\Users\86155\Downloads\智库研究 (1).xlsx'  # 输入的Excel文件路径
output_excel_path = r'C:\Users\86155\Desktop\学习\加勒比地区图库\output.xlsx'  # 插入图片后的新Excel文件路径

# 加载Excel文件
wb = load_workbook(excel_path)
ws = wb.active  # 获取活动工作表

# 获取图片链接列的列索引
header = [cell.value for cell in ws[1]]  # 获取第一行的所有列标题
image_links_column_index = header.index("图片链接") + 1  # 获取“图片链接”列的列号（Excel是1-based索引）

# 初始化图片计数器
image_counter = 1

# 获取Excel的最大列数
max_column = ws.max_column

# 遍历每个图片链接，下载并插入图片
for row in range(2, ws.max_row + 1):  # 从第二行开始处理
    image_url = None  # 确保在异常处理时 image_url 已定义
    try:
        image_url = ws.cell(row=row, column=image_links_column_index).value  # 获取图片链接列的值

        if image_url:  # 如果图片链接不为空
            # 为图片生成唯一的文件名（通过计数器）
            img_name = f"image_{image_counter}.jpg"  # 使用计数器作为文件名
            img_path = os.path.join(output_folder, img_name)

            # 请求图片内容
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'
            }
            img_response = requests.get(image_url, headers=headers)

            # 检查图片是否成功下载
            if img_response.status_code == 200:
                with open(img_path, 'wb') as f:
                    f.write(img_response.content)
                print(f"下载图片成功: {img_name}")

                # 创建图片对象
                img = Image(img_path)

                # 获取B列单元格宽度和当前行的高度
                column_width = ws.column_dimensions[get_column_letter(max_column + 1)].width  # 获取新列的宽度
                row_height = ws.row_dimensions[row].height  # 获取当前行的高度

                # 处理列宽和行高为None的情况，设置默认值
                if column_width is None:
                    column_width = 10  # 默认宽度
                if row_height is None:
                    row_height = 15  # 默认行高

                # 动态调整图片大小
                img_width = column_width * 7  # 图片宽度，适配Excel列宽
                img_height = row_height * 0.75  # 图片高度，适配Excel行高
                img.width = img_width
                img.height = img_height

                # 将图片插入到Excel的下一列
                img.anchor = f'{get_column_letter(max_column + 1)}{row}'  # 将图片插入到新列
                ws.add_image(img)
            else:
                print(f"图片下载失败: {image_url}, 状态码: {img_response.status_code}")
        else:
            print(f"没有图片链接，跳过第{row}行")

    except Exception as e:
        print(f"下载图片失败: {image_url}, 错误: {e}")

    # 每次处理完一个链接后，计数器增加
    image_counter += 1

# 保存修改后的Excel文件
wb.save(output_excel_path)
print(f"图片已插入到Excel，文件保存为 {output_excel_path}")
```

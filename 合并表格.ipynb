```python
import os
import pandas as pd

# 设置文件夹路径
folder_path = r'C:\Users\86155\Downloads\土库曼斯坦'

# 获取所有文件名
files = os.listdir(folder_path)

# 筛选出所有以 'all_data_part_' 开头并且是 CSV 文件的文件
csv_files = [file for file in files if file.startswith('all_data_part_') and file.endswith('.csv')]

# 读取所有的 CSV 文件并合并
all_data = pd.DataFrame()  # 初始化一个空的 DataFrame 用于存储合并后的数据

for csv_file in csv_files:
    # 构造文件的完整路径
    file_path = os.path.join(folder_path, csv_file)
    
    # 读取当前 CSV 文件
    df = pd.read_csv(file_path)
    
    # 合并到 all_data DataFrame
    all_data = pd.concat([all_data, df], ignore_index=True)
```

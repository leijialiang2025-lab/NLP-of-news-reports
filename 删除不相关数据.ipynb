```python
import pandas as pd
from openai import OpenAI
import time
import json

# 配置 OpenAI 客户端
client = OpenAI(
    base_url="https://chatapi.littlewheat.com/v1",
    api_key="sk-w33LIKFEXz7UHKg2R5tVvboSNAU76iCZsvyOyOjPtLLPp89S"  # 请替换为你的 API Key
)

# 读取 Excel 文件并截断数据
file_path = r"C:\Users\86155\Downloads\新浪网_sb.xlsx"
#out_path = r"C:\Users\86155\Downloads\新浪网_sb1.xlsx"
df = pd.read_excel(file_path)
empty_rows = df[df.isnull().all(axis=1)].index
first_empty_row = empty_rows[0] if not empty_rows.empty else len(df)
df = df.iloc[:1001]  # 截取前 500 行

# 调用 API 的函数
def call_gpt_api(message, max_token, retries):
    for _ in range(retries):
        try:
            response = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=message,
                max_tokens=max_token,
                timeout=30
            )
            return response.choices[0].message.content.strip()
        except Exception as e:
            print(f"API调用失败: {e}, 重试中...")
            time.sleep(1)
    return ""

# 定义分类问题
questions = """
请回答以下问题：
- 问题一：文本中的事件是否有与加勒比地区相关，请回答是或否
请严格返回 JSON 格式的结果，返回格式：{"问题一": "xxx"}
"""

# 分类函数
def Category(text):
    kinds = [
        {"role": "system", "content": "你是一个熟知加勒比地区的资深新闻资讯撰写人。"},
        {"role": "user", "content": f"通读新闻:{text}，{questions}"}
    ]
    response_content = call_gpt_api(kinds, 50, 3)
    try:
        answers = json.loads(response_content)
        return answers.get("问题一", "")
    except json.JSONDecodeError:
        print("问题回答过程出现问题")
        return "未知"

# 使用 xlsxwriter 保存并设置行颜色
with pd.ExcelWriter(file_path, engine="xlsxwriter") as writer:
    # 将 DataFrame 写入 Excel
    df.to_excel(writer, index=False, sheet_name="Sheet1")
    
    # 获取工作簿和工作表对象
    workbook = writer.book
    worksheet = writer.sheets["Sheet1"]
    
    # 定义黄色背景格式
    yellow_format = workbook.add_format({"bg_color": "#FFFF00"})
    
    # 遍历 DataFrame 的每一行
    for row_idx in range(len(df)):
        text = df.iloc[row_idx]["正文"]
        answers = Category(text)
        print(f"正在处理第{row_idx + 1}/{len(df)}条新闻...")
        
        # 如果条件不满足，设置整行黄色
        if answers != "是":
            # Excel 行号从 0 开始，标题行是 0，数据行从 1 开始
            excel_row = row_idx + 1  # 对应实际数据行
            worksheet.set_row(excel_row, None, yellow_format)
df.to_excel(file_path, index=False)
```
